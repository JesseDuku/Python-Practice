3. Create the 2 rotation matrices *R1* and *R2*: the first with an angle of 30[degree] and the second with an angle of 60[degree]. Print both matrices. **[0.5]**
4. Rotate the given point cloud by *R1* and save the result in *points_r1*. Be aware of the difference between a matrix multiplication and an elementwise multiplication. **[0.5]**
5. Rotate *points_r1* by *R2* and save the result in *points_r2*. **[0.5]**
6. Shift points_r2 by the *translation=(3,4)^T* and save the result in *points_t*. **[0.5]**
7. Create a figure with 4 subplots. Plot *points*, *points_r1*, *points_r2* and *points_t* in the created subplots, respectively. **[2]**

````python
# generating an L-shaped point cloud.
import numpy as np
np.random.seed(0) # For reproducable results
points = np.hstack([np.array([[5,0.5]]).T* np.random.rand(2,100),np.array([[0.5,5]]).T*np.random.rand(2,100)])
print('Point cloud dimension:',points.shape)
````

````python
# Plot points here
plt.scatter(points[0, :], points[1, :])
plt.title('Original Point Cloud')
plt.xlabel('X')
plt.ylabel('Y')
plt.axis('equal')
plt.show()
````

````python
# write angle2rotMat(...):
def angle2rotMat(angle_deg):

    angle_rad = np.radians(angle_deg)
    cos_val = np.cos(angle_rad)
    sin_val = np.sin(angle_rad)
    return np.array([[cos_val, -sin_val], [sin_val, cos_val]])
````

````python
# Create rotation matrices
R1 = angle2rotMat(30)
R2 = angle2rotMat(60)
print("Rotation Matrix R1:\n", R1)
print("Rotation Matrix R2:\n", R2)

# Rotate the point cloud
points_r1 = np.dot(R1, points)  # Rotate by R1
points_r2 = np.dot(R2, points_r1)  # Rotate points_r1 by R2


````

````python
# create translation vector

# apply Transformations
translation = np.array([3, 4])
points_t = points_r2 + translation[:, np.newaxis]
````

````python
# create subplots and plot results
plt.figure(figsize=(12, 12))

# Original points
plt.subplot(2, 2, 1)
plt.scatter(points[0, :], points[1, :])
plt.title('Original Points')
plt.axis('equal')

# Points after rotation R1
plt.subplot(2, 2, 2)
plt.scatter(points_r1[0, :], points_r1[1, :])
plt.title('Points after Rotation R1')
plt.axis('equal')

# Points after rotation R2
plt.subplot(2, 2, 3)
plt.scatter(points_r2[0, :], points_r2[1, :])
plt.title('Points after Rotation R2')
plt.axis('equal')

# Points after translation
plt.subplot(2, 2, 4)
plt.scatter(points_t[0, :], points_t[1, :])
plt.title('Points after Translation')
plt.axis('equal')

plt.show()

````

2. Plot the trajectory of the robot in world coordinates. **[2]**
    - The trajectory is computed by iteratively applying the measured odometry to the previous position:
$$ x_{i+1} = x_{i} + \delta x_{i},\qquad y_{i+1} = y_{i} + \delta y_{i}$$
3. Plot the trajectory of the robot in world coordinates from 07:30:00 to 8:00:00. **[1]**

````python
file_path = '/content/drive/MyDrive/Classroom/odom.data'
with open(file_path, 'r') as file:
    for _ in range(5):
        print(file.readline().strip())
````

````python
# 1.
import pandas as pd
import matplotlib.pyplot as plt

# File path
file_path = '/content/drive/MyDrive/Classroom/odom.data'

# Read the odometry data with the correct delimiter and format
odom_data = pd.read_csv(file_path, sep=',', header=None, names=['hours', 'minutes', 'seconds', 'delta_x', 'delta_y'])

# Convert hours, minutes, and seconds to a total seconds format for easy filtering
odom_data['total_seconds'] = odom_data['hours']*3600 + odom_data['minutes']*60 + odom_data['seconds']

# Filter data for the time range 07:30:00 (27000 seconds) to 08:00:00 (28800 seconds)
filtered_data = odom_data[(odom_data['total_seconds'] >= 27000) & (odom_data['total_seconds'] <= 28800)]

# Initialize the first position
x, y = [filtered_data.iloc[0]['delta_x']], [filtered_data.iloc[0]['delta_y']]

# Compute the trajectory
for i in range(1, len(filtered_data)):
    x.append(x[-1] + filtered_data.iloc[i]['delta_x'])
    y.append(y[-1] + filtered_data.iloc[i]['delta_y'])


````

````python
# 2.
plt.figure(figsize=(10, 6))
plt.plot(x, y, marker='o')
plt.title('Robot Trajectory from 07:30:00 to 08:00:00')
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.grid(True)
plt.show()
````

#### **F2. Robot navigation in grid world**
Lets assume we have given a robot which is moving in a 2D grid world. We want to visualize the parts of the map which got explored by the robot.

**Tasks:**
1. Generate a 5x5 grid world by using a numpy matrix where each cell is initially zero. **[0.5]**
2. A point robot navigates on this grid world starting from cell (0,0) that is in the top-left corner of the grid. Implement the navigation actions of the robot and execute the following sequence of commands: *down*, *right*, *up*, *right*, *right*, *down*, *down*, *down*, *left*. Print the final position of the robot. **[3.0]**
3. Mark the cells visited by the robot in 2. on the grid world by setting their value to $1$ and plot the grid world by using `matplotlib`. **[1.5]**

````python
# 1.
import numpy as np

# Create a 5x5 grid world
grid_world = np.zeros((5, 5))
````

````python
# 2.
# Define the starting position of the robot
robot_position = [0, 0]  # (row, column)

# Sequence of commands
commands = ['down', 'right', 'up', 'right', 'right', 'down', 'down', 'down', 'left']

# Function to move the robot based on the command
def move_robot(position, command):
    if command == 'down':
        position[0] += 1
    elif command == 'up':
        position[0] -= 1
    elif command == 'right':
        position[1] += 1
    elif command == 'left':
        position[1] -= 1
    return position

# Execute the sequence of commands
for command in commands:
    robot_position = move_robot(robot_position, command)
    # Mark the visited cell
    grid_world[robot_position[0], robot_position[1]] = 1

# Print the final position of the robot
print("Final position of the robot:", robot_position)
````

````python
# 3.
import matplotlib.pyplot as plt

# Plot the grid world
plt.imshow(grid_world, cmap='Greys', interpolation='nearest')
plt.title('Robot Path in Grid World')
plt.xlabel('Column')
plt.ylabel('Row')
plt.grid(which='major', axis='both', linestyle='-', color='k', linewidth=2)
plt.xticks(range(5))
plt.yticks(range(5))
plt.show()
````
